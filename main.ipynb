```python
import pandas as pd
import numpy as np
```

```python
#Lendo o arquivo csv
df = pd.read_csv("data/Car_Prices_Poland_Kaggle.csv", encoding='utf8', sep=",", index_col=0)
df.head()
df.price = np.int32(df.price)
```

```python
#Carro com o preço mais alto
df[df.price == df.price.max()]
```

```python
#Carro com o preço mais baixo
df[df.price == df.price.min()]
```

```python
#Média dos preços
df.mean()
```
